```python
#WAP to check if a specified value is contained ina  group of values
def is_group_member(group_data, n):
    for value in group_data:
        if n ==value:
            return True
    return False
    
```

```python
a = int(input("Enter the number:"))
print(is_group_member([5,8,3,4,6], a)) # ([list], value to be searched)
```

```python
#combining 2 DataFrames
Student_Data = {'Student_id': ['AAA-123-2211','BBB-542-8542','CCC-854-8541','AAA-856-7458',
                               'DDD-745-7841'], 'First_Name': ['Sam','Jim','Rick','Ben','Kevin'],
               'Last_Name': ['Edison','Mon','Dany','Denil','Pocket']}
student = pd.DataFrame(Student_Data, columns = ['Student_id','First_Name','Last Name'])
student
```

```python
Student_Course= {'stu_id':['AAA-123-2211','BBB-542-8542','CCC-854-7458','AAA-856-7458',
                               'DDD-745-7841'],'course':['ML','DS','RPA','BI','DL'],
                 'duration':[3,4,2,5,2]}
course = pd.DataFrame(Student_Course)
course
```

```python
print(pd.concat([student,course])) #pd.concat([first dataframe, 2nd dataframe)
```

```python
#index joining
print(pd.concat([student,course], axis =1)) #axis = 1 makes the joining based on index. Index is the number 0 1 2 on the left most row
```

```python
#in the above number 2 is incorrect as student id is diff
#merge
pd.merge(student,course,left_on='Student_id',right_on ='stu_id',how='inner') #inner joins only adds those values that are common between both tables
```

```python
pd.merge(student,course,left_on='Student_id',right_on='stu_id',how='left') #fills all rows on the left dataframe while making right empty(NaN) if needed
```

```python
pd.merge(student,course,left_on='Student_id',right_on='stu_id',how='right') #on right, all right row for join will be filled but the left part may be left NaN
```

```python
pd.merge(student,course,left_on='Student_id',right_on='stu_id',how='outer') #no cross matching is done, all values are simply put into the dataframe
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
    'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(left)
print('\n')
print(right)
```

```python
pd.merge(left,right,on=['id','subject_id']) #since the 2 records are given, so id and subject id should match for both cases
#inner join taken by default
```

```python
pd.merge(left,right,on=['id','subject_id'],how='left') #left join needs how='left' to be specified
```

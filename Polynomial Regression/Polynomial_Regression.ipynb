```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.metrics import r2_score
```

```python
os.chdir('<filepath>')
```

```python
data = pd.read_csv('Height_weight_Dataset.csv')
display(data)
```

```python
x = data.iloc[:,0:1].values
print(x.shape)
print(x)
```

```python
y = data.iloc[:,1].values
print(y.shape)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
print(x.shape)
print(y.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
print(lr)
```

```python
y_pred = lr.predict(x)
accuracy = r2_score(y,y_pred)
print(accuracy)
print(y_pred)
```

```python
result = pd.concat([data,pd.DataFrame(y_pred, columns = ['Prediction'])],axis = 1)
result['Difference'] = result['Height'] - result['Prediction']
print(result)
```

```python
plt.scatter(x,y, color = 'green')
plt.plot(x,y_pred, color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynom2 = PolynomialFeatures(degree = 2)
x_polynom = polynom2.fit_transform(x)
print(pd.DataFrame(x_polynom).head(50))
```

```python
PolyReg2 = LinearRegression() 
PolyReg2.fit(xpoly_train, ypoly_train)
print (PolyReg2)
```

```python
ypoly_pred2 = PolyReg2.predict(x_polynom)
accuracy2 = r2_score(ypoly_pred2,y)
print (accuracy2)
```

```python
plt.scatter(x,y, color = 'green') 
plt.plot(x, ypoly_pred2, color = 'blue') 
plt.title('Polynomial Regression') 
plt.xlabel('Age') 
plt.ylabel('Height') 
plt.show()
```

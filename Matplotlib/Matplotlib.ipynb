```python
#area graph
days =   [1,2,3,4,5]
sleep =  [7,6,4,5,8]
eat =    [2,3,4,2,1]
work =   [8,9,10,8,9]
play=    [4,5,3,3,2]
plt.stackplot (days,sleep,eat,work,play,labels= ('sleep','eat','work','play'), #plt.stackplot(parameters) used for area graph
               colors= ['m','r','k','g'])               
plt.title ('Graph Five')
plt.xlabel ('X- Axis')
plt.ylabel ('Y-Axis')
plt.legend()
plt.show()
```

```python
#pie chart
hours = [5,8,3,4] #creates in anti clockwise, can plot time/value spent of a total
plt.pie(hours) #plt.pie(array name)
plt.show()
```

```python
hours = [5,8,3,4]
activity = ['Sleep', 'Work', 'Eat', 'Play'] #labels
col = ['g','r','c','y']
plt.pie(hours,labels = activity, colors = col, startangle = 90, shadow = True, explode = (0,0,0,0.1),autopct = '%1.1f%%')
# explode makes one of the given values pop out in anti clockwise order, autopct gives a decimal point
plt.title('Pie Chart')
plt.axis('equal') #makes equal aspect ratio 1:1 to ensure that chart is circular
plt.show()
```

```python
#donut chart
plt.figure(figsize= (5,5))
hours = [5,8,3,4]
plt.pie (hours)
inner_circle = plt.Circle((0,0), 0.70, fc = 'w') #plt.circle((x,y), width, facecolor i.e. color of the interior circle
plt.gca().add_artist(inner_circle)
plt.show()
```

```python
#double graphs in one
x1 = np.arange(10,30,2)
y1 = [1,5,6,7,8,3,4,6,5,4]
x2 = np.arange(10,30,2)
y2 = [5,4,2,8,4,6,3,4,5,4]
plt.plot(x1,y1) #gives first plot with x and y coords of points
plt.plot(x2,y2) #gives second plot with x and y coords of points
plt.show()
```

```python
#seperating the 2 graphs
x1=np.arange(10,30,2)
y1 = [1,5,6,7,8,3,4,6,5,4]
x2= np.arange (10,30,2)
y2 = [5,4,2,8,4,6,3,4,5,4]
plt.subplot(221) #the first and second 2 shows the dimensions of split which is 2x2 here, the 1 at end is inded of graph
plt.plot(x1,y1)
plt.subplot(222) #the index here is the last '2'. This means that this graph will be plotted after the previous one
plt.plot(x2,y2)
plt.show()
```

```python
x1=np.arange(10,30,2)
y1 = [1,5,6,7,8,3,4,6,5,4]
x2= np.arange (10,30,2)
y2 = [5,4,2,8,4,6,3,4,5,4]
plt.subplot (211)
plt.plot(x1,y1)
plt.title ('Graph 1')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')
plt.subplot(212)
plt.plot(x2,y2)
plt.title ('Graph 2')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')
plt.show()
```

```python
data = pd.read_csv('Matplotlib_data.csv') #path not needed as directory changed to current directory
data
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444', #color takes the color of line through hexcode
         linestyle='--', label='All Devs') #linestyle gives style of line
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips1 = tips
print (tips1)
```

```python
sns.boxplot (data = tips1 , x = 'day', y = 'total_bill' )
plt.show()
```

```python
#In the above boxplot, the lowest line is the lowest value, the line at bottom of plot color is 25%, line in middle is 50%
#line in top is 75% and top most line is 75% and top most line is max value. Any data above that is outlier
```

```python
tips.describe() #gives a writen format of the above boxplot
```

```python
print(tips[tips1['day'] == 'Fri'].describe()) #describes for fridat only
```

```python
#displot: distribution plot
sns.displot(tips [tips1['day'] =='Fri']['total_bill'], kde = True)
plt.show()
```

```python
sns.displot(tips1['total_bill'], kde = True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
os.chdir('<filepath>')
```

```python
df1 = pd.read_csv('Social_Network_Ads.csv')
print(df1)
```

```python
x = df1.iloc[:,2:4].values
print(x)
y = df1.iloc[:,4].values
print(y)
```

```python
plt.scatter (x[y == 0,0],x[y == 0,1] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y == 1,0],x[y == 1,1] ,s = 100, c='b', label = 'Cluster 2')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
print(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
print (x.shape)
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
#Gaussian NB model
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(x_train,y_train)
print(GNB)
```

```python
GNB_pred = GNB.predict(x)
print(GNB_pred)
```

```python
df_GNBpred = pd.DataFrame(GNB_pred,columns=['Predictions'])
result = pd.concat([df1, df_GNBpred],axis = 1)
print(result)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y,GNB_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report (y, GNB_pred, zero_division=0))
```

```python
cm = confusion_matrix (y , GNB_pred)
print (cm)	
```

```python
cm = confusion_matrix (y , GNB_pred)
print (cm)	
```

```python
from sklearn.naive_bayes import BernoulliNB
BNB = BernoulliNB()
BNB.fit(x_train,y_train)
print (BNB)
```

```python
BNB_pred = BNB.predict(x)
print(BNB_pred)
```

```python
df_BNBpred = pd.DataFrame(BNB_pred, columns = ['Prediction'])
result = pd.concat([df1, df_BNBpred], axis = 1)
print(result)
```

```python
cm = confusion_matrix(y,BNB_pred)
print(cm)
```

```python
#MultonomialNB: All data should be positive number(limitation)
#so min max scaler is used to put all data between 0 and 1
x= df1.iloc [:,2:4].values
y= df1.iloc[:,4].values
from sklearn.preprocessing import MinMaxScaler
minmax_x= MinMaxScaler()
minmax_x = minmax_x.fit_transform(x)
print (minmax_x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(minmax_x,y,test_size= 0.2)
print (x.shape)
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(x_train,y_train)
print(MNB)
```

```python
MNB_pred = MNB.predict(x)
print(MNB_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report (y,MNB_pred,zero_division=0))
```

```python
cm = confusion_matrix (y, MNB_pred)
print (cm)	
```
